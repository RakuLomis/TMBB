```python
def convert_timestamps(timestamp): 
    # 分割两个时间戳
    parts = timestamp.split('--') 
    
    # 解析并格式化每个时间戳
    dt1 = datetime.strptime(parts[0] + '--' + parts[1], '%y-%m-%d--%H-%M-%S') 
    dt2 = datetime.strptime(parts[0] + '--' + parts[2], '%y-%m-%d--%H-%M-%S') 
    
    # 格式化为目标字符串
    formatted_timestamp1 = dt1.strftime('20%y-%m-%dT%H:%M:%S.0000000+08:00') 
    formatted_timestamp2 = dt2.strftime('20%y-%m-%dT%H:%M:%S.0000000+08:00') 
    
    return formatted_timestamp1, formatted_timestamp2 
```

```python
# 输入的时间戳
timestamp = '24-09-26--17-03-48--20-04-24'

# 转换并输出
formatted_timestamp1, formatted_timestamp2 = convert_timestamps(timestamp)
print(formatted_timestamp1)
print(formatted_timestamp2)
```

```python
import math

def cosine_test(list1, list2): 
    dot = 0
    num1 = 0
    num2 = 0
    for i in range(len(list1)): 
        dot += list1[i] * list2[i] 
        num1 += list1[i] * list1[i] 
        num2 += list2[i] * list2[i] 
    num1 = math.sqrt(num1) 
    num2 = math.sqrt(num2) 
    return dot / (num1 * num2)

res = cosine_test(vector1, vector2)
```

```python
import numpy as np

def l2_normalize(vector):
    norm = np.linalg.norm(vector)
    return vector / norm

def normalized_euclidean_distance(v1, v2):
    v1_normalized = l2_normalize(v1)
    v2_normalized = l2_normalize(v2)
    return np.linalg.norm(v1_normalized - v2_normalized)

ip1 = [192, 168, 5, 2, 120, 232, 217, 159]
ip2 = [198, 18, 0, 1, 198, 18, 0, 161]

distance = normalized_euclidean_distance(ip1, ip2)
print(f'Normalized Euclidean Distance: {distance}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 创建 DataFrame
data = {
    'Index': [1, 4, 5, 100],
    'W Flows': [4, 5, 9, 2]
}
df = pd.DataFrame(data)

# 绘制柱状图
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(df)), df['W Flows'], color='skyblue')

# 在柱的顶部添加 '*' 标记
plt.plot(range(len(df)), df['W Flows'], marker='*', color='orange', linestyle='-')

# 设置 x 轴刻度和标签
labels = df['Index'].astype(str).tolist()
# 在大间隔处添加省略号
for i in range(1, len(labels)):
    if df['Index'][i] - df['Index'][i-1] > 10:  # 自定义间隔判断
        labels.insert(i, '...')

plt.xticks(range(len(labels)), labels)

# 添加标题和标签
plt.title('Bar Chart with Line Plot')
plt.xlabel('Index')
plt.ylabel('W Flows')

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


plt.rcParams['font.family'] = 'Microsoft YaHei'
# 数据
w_flows = [1, 1, 7, 1, 2, 2, 1, 1, 1, 2, 1]

# 绘制直方图
plt.figure(figsize=(4, 3))
count, bins, ignored = plt.hist(w_flows, bins=range(min(w_flows), max(w_flows) + 2), 
                                color='skyblue', edgecolor='black', density=True, align='left')

# 拟合正态分布曲线
mu, std = norm.fit(w_flows)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# 绘制拟合曲线
plt.plot(x, p, 'orange', linewidth=2)

# 添加标题和标签
plt.title('Histogram and Normal Distribution Fit')
plt.xlabel('W Flows')
plt.ylabel('Density')

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 设置字体
plt.rcParams['font.family'] = 'Times New Roman'

# 两组数据
data1 = [1, 1, 7, 1, 2, 2, 1, 1, 1, 2, 1]
# data2 = [2, 2, 8, 2, 3, 3, 2, 2, 2, 3, 2]
data2 = [1, 1, 7, 1, 2, 2, 1, 1, 1, 2, 1]

# 计算直方图数据
bins = range(min(min(data1), min(data2)), max(max(data1), max(data2)) + 2)
hist1, bins1 = np.histogram(data1, bins=bins, density=True)
hist2, bins2 = np.histogram(data2, bins=bins, density=True)

# 设置偏移量
width = 0.4
offset = width / 2

# 绘制直方图
plt.figure(figsize=(10, 6))
plt.bar(bins1[:-1] - offset, hist1, width=width, color='skyblue', edgecolor='black', label='Data 1', align='edge')
plt.bar(bins2[:-1] + offset, hist2, width=width, color='lightcoral', edgecolor='black', label='Data 2', align='edge')

# 拟合正态分布曲线 - 数据1
mu1, std1 = norm.fit(data1)
x1 = np.linspace(min(data1), max(data1), 100)
p1 = norm.pdf(x1, mu1, std1)
plt.plot(x1, p1, 'blue', linewidth=2)

# 拟合正态分布曲线 - 数据2
mu2, std2 = norm.fit(data2)
x2 = np.linspace(min(data2), max(data2), 100)
p2 = norm.pdf(x2, mu2, std2)
plt.plot(x2, p2, 'red', linewidth=2)

# 添加标题和标签
plt.title('Histograms and Normal Distribution Fits')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# 显示图表
plt.tight_layout()
plt.show()
```
